# Machine learning in Python

## Linear curve fitting

````python
from scipy.optimize import curve_fit
````

````python
def f(x,a,b):
    return a*x+b
````

````python
x=linspace(0,10,100)
y=x+random.randn(100) # random from numpy
popt,pcov=curve_fit(f,x,y)
a,b=popt
popt
````

````python
scatter(x,y)
plot(x, a*x + b, 'r',label="estimate")
ax=gca()
ax.legend();
````

## Interpolation

````python
from scipy.interpolate import interp1d
````

````python
x1=linspace(0,10,10)
y1=x1*sin(x1)
scatter(x1,y1,s=30)
# interpolate
f=interp1d(x1, y1, kind='cubic')
x2=linspace(0,10,100)
plot(x2,x2*sin(x2),'--')
plot(x2,f(x2),'r')
````

## Classification

````python
from sklearn.svm import SVC
````

````python
X = array([[-1, -1], [-2, -1], [1, 1], [2, 1]])
Y = array([1, 1, 2, 2])
````

````python
clf = SVC()
clf.fit(X,Y)
````

````python
clf.predict([[-0.8,-1]])
````

````python
scatter(X[:,0], X[:,1],s=50,c=Y)
````

````python
gd = array([[i,j] for i in arange(-2.5, 2.5, 0.2) for j in arange(-2.5, 2.5, 0.2)])
gdc = clf.predict(gd)
scatter(gd[:,0], gd[:,1], s=50, c=gdc);
````
